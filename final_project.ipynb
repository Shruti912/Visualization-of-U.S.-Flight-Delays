## Method 1: Interactive Cancellation & Diversion Dashboard

What I Did:
With the dropdowns for year, airport and month, I have constructed this little interactive dashboard. Once you select something it immediately triggers the data filter, it will pull up the total number of canceled and diverted flights by selecting the dataset and then aggregate the results.

Why I Chose This Method:
Cancellations and diversions are a spikey thing and vary widely based on the time and place. With all those airport - month combos out there, it's good to have an interactive tool where we can mine in specifically into situations and not be overwhelmed by the information for a simple interactive dataset.

Why This Works:
Panel widgets are updated on your fly when you change a selection, so that you have fast and responsive exploration and the heavy lifting is done under the hood.


```python
DELAY_FACTORS = [
    "carrier_delay",
    "weather_delay",
    "nas_delay",
    "security_delay",
    "late_aircraft_delay"
]
df['total_delay'] = df[DELAY_FACTORS].sum(axis=1)
```

```python
# Create dropdown widgets
year_select = pn.widgets.Select(
    name='Select Year', 
    options=sorted(df['year'].unique())
)
airport_select = pn.widgets.Select(
    name='Select Airport', 
    options=sorted(df['airport_name'].unique())
)
month_select = pn.widgets.Select(
    name='Select Month', 
    options=sorted(df['month'].unique())
)

# Bind display function to widgets using our module function
summary_panel = pn.bind(
    display_counts, 
    df=df,
    year=year_select, 
    airport=airport_select, 
    month=month_select
)
```

```python
# Create dashboard
dashboard = pn.Column(
    pn.pane.Markdown("## Airline Cancellation & Diversion Dashboard"),
    pn.Row(
        pn.Column(year_select, airport_select, month_select)
    ),
    summary_panel
)

dashboard
```

## Method 2: Combo Chart  Total Delay vs. Selected Delay Factor

What I Did:
I created a combo chart of a bar chart and a line chart to compare delay in total and for user-chosen delay cause for the top ten airlines or airports. Bars indicate the total delay; the line displays a zooming in of a certain factor's contribution.

Why I Chose This Method:
Just this, looking at total delay misses the real reasons behind it. Overlaying a specific cause allows you to understand how the big picture compares to the tops, all in the same visual.

Why This Works:
Using two visual channels- bar height and line slope- the chart can be read and attention can be paid to the largest contribution players.

```python
ENTITY_OPTIONS = {
    "Airlines": ("carrier_name", "carrier"),   
    "Airports": ("airport_name", "airport")   
}

# creating widgets 
entity_select = pn.widgets.Select(
    name="Select Category",
    options=list(ENTITY_OPTIONS.keys()),
    value="Airlines"
)
factor_select = pn.widgets.Select(
    name="Select Delay Factor",
    options=DELAY_FACTORS,
    value="carrier_delay"
)
```

```python
#binding combo chart function
dashboard_combo = pn.bind(
    compute_combo_chart,
    df=df,
    entity=entity_select,
    factor=factor_select,
    entity_options=ENTITY_OPTIONS,
    delay_factors=DELAY_FACTORS
)
combo_dashboard = pn.Column(
    pn.pane.Markdown("## Chart  Total Delay + Selected Delay Factor"),
    pn.Row(entity_select, factor_select),
    dashboard_combo
)
combo_dashboard
```

### Merging Airport Coordinates

Geospatial analysis requires accurate airport location data. Unfortunately the data we downloaded didnt has Longitude and Latitude data so we individually searched airport specific data and culminated a new data set for every airport in correspondence to its location. Later on cleaning and merging it provide a better and outcome oriented csv file. 

The resulting dataset includes both delay metrics and precise airport locations, making it suitable for spatial visualization and geographic analysis.

```python
df_coords = pd.read_csv('./data/airports_lon_lat.csv')
df_coords_clean = df_coords[['IATA Code', 'latitude', 'longitude']]
df_merged_final = pd.merge(
    df,
    df_coords_clean,
    left_on='airport',
    right_on='IATA Code',
    how='left'
)
df_merged_final.drop(columns=['IATA Code'], inplace=True)
df_merged_final.head()
```

## Method 3: Geospatial Heatmap of Average Airport Delays

What I Did:  
I combined the delay data with the latitude and longitude coordinates, and so I calculated a composite delay metric based on the average of the five cause categories. Airports were placed over an OpenStreetMap base layer, color intensity and point size encoders were used to represent magnitude of delay.  

Why I Chose This Method:  
Delays are greatly affected by geography - traffic congestion, weather patterns and the underlying network. By featuring them onto a map we immediately identify spatial patterns that are not obtainable by tables or simple plots.  

Why This Works:  
The heatmap shows regional delay hot spots immediately, and interactive tooltips enable you to click on an airport to get the exact numbers at times when you may need to see more details.  


```python
df_grouped = df.groupby('airport')[delay_cols].sum()
df_grouped['avg_delay_cause'] = df_grouped.mean(axis=1)
df_grouped = df_grouped.reset_index()
```

```python
df_map_data = pd.merge(
    df_grouped,
    df_coords_clean,
    left_on='airport',
    right_on='IATA Code',
    how='left'
)
points = gv.Points(
    df_map_data, 
    kdims=['longitude', 'latitude'], 
    vdims=['avg_delay_cause', 'airport']
).opts(
    title="Average Delay by Airport",
    color='avg_delay_cause',
    cmap='Inferno_r',           
    size=gv.dim('avg_delay_cause').norm() * 20 + 5,  
    alpha=0.8,
    colorbar=True,
    clabel='Avg Delay (Minutes)',
    tools=['hover'],
    width=900,
    height=500
)
```

```python
tiles = gts.OSM()
heatmap = tiles * points
heatmap
```

# Method 4: Monthly Airline Delay by Cause for specific Airport

What we did?
We analyzed flight delay data for selected airlines operating at specific airports. The delays were aggregated and visualized to compare performance across airlineairport combinations.


Why did we use this?
This analysis was used to examine how flight delays vary for specific airlines at individual airports. It helps compare airline performance across locations and identify recurring delay patterns. These insights support a better understanding of operational inefficiencies.

```python
df[delay_cols] = df[delay_cols].fillna(0)
df['month'] = df['month'].astype(str)
        


df_melted = df.melt(
    id_vars=['year', 'carrier_name', 'airport', 'month'],
    value_vars=delay_cols,
    var_name='Delay Type',
    value_name='Minutes'
)

name_map = {
    'carrier_delay': 'Carrier',
    'weather_delay': 'Weather',
    'nas_delay': 'NAS',
    'security_delay': 'Security',
    'late_aircraft_delay': 'Late Aircraft'
}
```

```python
ds = hv.Dataset(
    df_melted,
    kdims=['month', 'Delay Type', 'year', 'carrier_name', 'airport'],
    vdims=['Minutes']
)
```

```python
from flight_analysis import error_check
```

```python

bars = ds.groupby(
    ['year', 'carrier_name', 'airport'],
    dynamic=True
).apply(error_check)

styled_plot = bars.opts(
    opts.Bars(
        stacked=True,
        width=600,
        height=400,
        title="Monthly Airline Delays by Cause for Specific Year and Airport",
        tools=['hover'],
        xlabel='Month',
        ylabel='Delay Minutes',
        legend_position='right',
        xrotation=45,
        framewise=True,
        shared_axes=False
    )
)

styled_plot
```
